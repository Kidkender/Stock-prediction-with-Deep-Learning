```python
df = DataReader('AAPL', data_source='yahoo', start='2010-01-01', end='2021-01-01')
df
```

```python
plt.figure(figsize=(16,6))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.show()
```

```python
data = df.filter(['Close'])
dataset = data.values
training_data_len = int(np.ceil( len(dataset) * .95 ))

training_data_len
```

```python
len(dataset)
```

```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python

train_data = scaled_data[0:int(training_data_len), :]
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print()
        
x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
from keras.layers.core import Dense, Dropout, Activation, Lambda,Flatten
from keras.layers.convolutional import Convolution1D
from keras.layers import Conv1D, Dense, MaxPool1D, Flatten, Input,GlobalMaxPooling1D
import time
```

```python
#Step 2 Build Model
model_LSTM1 = Sequential()

model_LSTM1.add(LSTM(units=50,input_dim=1,return_sequences=True))
model_LSTM1.add(Dropout(0.2))

model_LSTM1.add(LSTM(100,return_sequences=False))
model_LSTM1.add(Dropout(0.2))

model_LSTM1.add(Dense(units=1))
model_LSTM1.add(Activation('linear'))

start = time.time()
model_LSTM1.compile(loss='mse', optimizer='rmsprop',metrics=['accuracy'])
print ('compilation time : ', time.time() - start)
```

```python
model_LSTM1.fit(
    x_train,
    y_train,
    batch_size=128,
    epochs=10,
    validation_split=0.05)
```

```python
# Build the LSTM model
model_LSTM2 = Sequential()
model_LSTM2.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model_LSTM2.add(LSTM(64, return_sequences=False))
model_LSTM2.add(Dense(25))
model_LSTM2.add(Dense(1))

# Compile the model
model_LSTM2.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
```

```python
model_LSTM2.fit(x_train, y_train, batch_size=128, epochs=10)
```

```python
model_ANN1 = Sequential()
model_ANN1.add(Dense(128, input_dim=(x_train.shape[1]), activation='relu'))
model_ANN1.add(Dense(128, activation='relu'))
model_ANN1.add(Dense(64, activation='relu'))
model_ANN1.add(Dense(16, activation='relu'))
model_ANN1.add(Dense(8, activation='relu'))
model_ANN1.add(Dense(1, activation='softmax'))
model_ANN1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_ANN1.fit(x_train, y_train, epochs=10, batch_size=128)
```
